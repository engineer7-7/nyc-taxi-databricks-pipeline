- import the parquet dataset

```python
# initialize the file path and the table name which it will be named in the bronze db
configs = {
    '/Volumes/workspace/taxi_project/raw_sources/yellow_tripdata_2025-01.parquet' : 'bronze_yellow_taxi_trips_2025_01'
}
```

```python
# function to read the parquet file

def read_parquet(filename):
  return spark.read.parquet(filename)
```

```python
# function to store the df in the bronze db

def parquet_to_table(dataframe, tablename):
    return(dataframe.write
     .format("delta")
     .mode("overwrite")  
     .saveAsTable(f"bronze.{tablename}")
    )
```

```python
for filename, tablename in configs.items():
    df = read_parquet(filename=filename)
    parquet_to_table(dataframe=df, tablename=tablename)
```
