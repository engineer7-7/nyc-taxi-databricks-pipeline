```python
df_gold_trips_daily = spark.sql(
    """
    WITH cte_duration AS (SELECT gold_taxi_nyc.*,
                             TIMESTAMPDIFF(MINUTE, pickup_datetime, dropoff_datetime) AS duration_minutes
                        FROM workspace.silver.yellow_taxi_trips_2025_01 gold_taxi_nyc)
    SELECT trip_date,
       COUNT(*)                        AS trips_count,
       ROUND(SUM(total_amount), 2)     AS revenue_total,
       ROUND(AVG(total_amount), 2)     AS avg_total_amount,
       ROUND(AVG(trip_distance), 2)    AS avg_trip_distance,
       ROUND(AVG(duration_minutes), 2) AS avg_trip_duration,
       ROUND(AVG(fare_amount), 2)      AS avg_fare_amount
FROM cte_duration
GROUP BY trip_date;
    """
)
```
